```python
from policyengine.models import *
from policyengine.database import Database

db = Database(url="postgresql://postgres:postgres@127.0.0.1:54322/postgres")
```

```python
current_law = db.get(Policy, name="Current law", country="uk")
static = db.get(Dynamic, name="Static", country="uk")
frs_2029 = db.get(Dataset, name="frs_2023_24/2029")
```

```python
sim = Simulation(
    policy=current_law,
    dynamic=static,
    dataset=frs_2029,
    country="uk",
    model_version="2"
)
```

```python
sim.run()
```

```python
db.add(sim)
```
